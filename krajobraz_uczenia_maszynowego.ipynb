```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression
from pathlib import Path 

np.random.seed(42)

plt.rc('font', size=12)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=12)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)
```

```python
import urllib.request
datapath = Path() /"data"
oecd_bli = pd.read_csv(datapath/"D:\Bartlomiej Ryczek\wum_is2\data\oecd_bli.csv") 
gdp_per_capita=pd.read_csv(datapath/"D:\Bartlomiej Ryczek\wum_is2\data\gdp_per_capita.csv")
```

```python
display(oecd_bli)
```

```python
gdppc_col = "GDP per capita (USD)"
lifesat_col = "Life satisfaction"

gdp_per_capita_2020 = gdp_per_capita[gdp_per_capita["Year"] == 2020]
gdp_per_capita = gdp_per_capita.drop(["Code", "Year"], axis=1)
gdp_per_capita.columns = ["Country", gdppc_col]
gdp_per_capita.set_index("Country", inplace=True)

gdp_per_capita.head()
```

```python
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
oecd_bli
```

```python
oecd_bli.reset_index()
```

```python
full_country_stats = oecd_bli.merge(gdp_per_capita_2020, left_on="Country", right_on="Entity", how="inner")
full_country_stats
```

```python

full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                              left_index=True, right_index=True)
full_country_stats.sort_values(by=gdppc_col, inplace=True)
full_country_stats = full_country_stats[[gdppc_col, lifesat_col]]

full_country_stats.head()
```

```python


full_country_stats = full_country_stats(["Country","Life satisfaction", "GDP per capita (USD)"])
full_country_stats
```

```python

full_country_stats.to_csv(datapath/"lifesat_full.csv", index=False)
```

```python
min_gdp = 23_500
max_gdp = 62_500

country_stats = full_country_stats [([full_country_stats["GDP per capita (USD)"]>=min_gdp])&([full_country_stats["GDP per capita (USD)"]<=max_gdp])]

country_stats.to_csv(datapath / "lifesat.csv")
```
